# metabeta — simulation-based diagnostics

This notebook validates **model quality** using simulated datasets where the ground-truth
parameters are known.  It is the companion to `intro.ipynb`, which covers the real-data API
workflow.

Because we know the truth, we can go beyond posterior summaries and run proper calibration
checks — things that are impossible on a single real dataset.

---

## Planned sections

### 1. Simulate a batch of datasets
- Instantiate `Generator` with a data config (toy YAML) to produce a batch of hierarchical
  datasets on the fly.
- Each dataset comes with known `ffx`, `sigma_rfx`, and `rfx` ground-truth arrays.
- Show a few example trajectories and the distribution of dataset sizes.

### 2. Load the model and run batch inference
- Load `mb = Router(JOINT_PATH)` from an existing joint checkpoint.
- Call `mb.sample()` on the full batch at once; print routing summary and sample shapes.

### 3. Parameter recovery
- `plotRecovery`: scatter posterior means vs. true values for `ffx` and `sigma_rfx` across
  all simulated datasets.
- A tight cloud around the diagonal confirms good point-estimate performance.

### 4. Posterior interval coverage
- `plotCoverage`: for each credible-interval level α, the fraction of datasets whose true
  parameter falls inside the α-CI.
- A calibrated model tracks the identity line; systematic over/under-coverage reveals
  bias or over-confidence.

### 5. Simulation-based calibration (SBC)
- `plotSBC`: rank-histogram of where the true parameter falls within the sorted posterior
  samples.
- Uniform histograms confirm a valid posterior approximation; U-shapes or humps indicate
  systematic bias or over/under-dispersion.

### 6. Predictive checks
- `plotPPD` on a selection of datasets, overlaying observed vs. posterior-predictive
  density envelopes.
- Optionally compare prior predictive vs. posterior predictive to visualise how much the
  data contracts the predictive distribution.
